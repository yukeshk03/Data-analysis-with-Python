```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

def update_layout(fig, title, xaxis_title=None, yaxis_title=None, height=300, width=450):
    fig.update_layout(
        plot_bgcolor='#f0f8ff',
        paper_bgcolor='#f0f8ff',
        title_font_color='#0056b3',
        title=title,
        font=dict(
            family='Arial',
            size=12,
            color='#0056b3'
        ),
        height=400,
        width=500
    )
    
    if xaxis_title:
        fig.update_xaxes(title_text=xaxis_title)
    if yaxis_title:
        fig.update_yaxes(title_text=yaxis_title)
    return fig

fig_age = update_layout(px.histogram(df, x='age'), 'Distribution of Player Ages', xaxis_title='Age', yaxis_title='Count')

fig_goals_position_spain = update_layout(px.pie(spain_df, names='position', values='goals', title='Goals by Position for Spain'), 'Goals by Position for Spain')
fig_goals_position_spain.update_traces(textposition='inside', textinfo='percent+label')

fig_goals_position_england = update_layout(px.pie(england_df, names='position', values='goals', title='Goals by Position for England'), 'Goals by Position for England')
fig_goals_position_england.update_traces(textposition='inside', textinfo='percent+label')

fig_market_caps = update_layout(px.scatter(df, x='caps', y='marketvalue', color='country', size='goals', title='Market Value vs. Caps'), 'Market Value vs. Caps', xaxis_title='Caps', yaxis_title='Market Value')

fig_goals_by_team = update_layout(px.bar(sefff, x='Team A', y='Goals', title='Goals by Country - Euro 2024'), 'Goals by Country - Euro 2024', xaxis_title='Country', yaxis_title='Goals')

fig_bubble = update_layout(px.scatter(sefff, x='Shots', y='Shots_on_target', size='Goals', color='Team A', title='Shots vs Shots on Target with Goals - Euro 2024'), 'Shots vs Shots on Target with Goals - Euro 2024', xaxis_title='Shots', yaxis_title='Shots on Target')

fig_goals_conceded = update_layout(px.bar(sefff, x='Team A', y='goalo', title='Goals Conceded by Country - Euro 2024'), 'Goals Conceded by Country - Euro 2024', xaxis_title='Country', yaxis_title='Goals Conceded')

fig_passes = update_layout(px.bar(sefff, x='Team A', y=['Passes', 'oppon_Passes'], barmode='group', title='Passes vs Opponent Passes - Euro 2024'), 'Passes vs Opponent Passes - Euro 2024', xaxis_title='Country', yaxis_title='Passes')

app.layout = html.Div([
    html.H1('EURO CUP FINAL EDA', style={
        'color': '#0056b3', 
        'font-family': 'Arial, sans-serif', 
        'text-align': 'center', 
        'font-size': '36px', 
        'font-weight': 'bold'
    }),
    html.Br(),  
    dbc.Tabs([
        dbc.Tab(label='Players List', children=[
            html.Div([
                html.Div([
                    dcc.Dropdown(
                        id='team-dropdown',
                        options=[
                            {'label': 'Spain', 'value': 'spain'},
                            {'label': 'England', 'value': 'england'}
                        ],
                        value='spain',
                        style={'color': '#0056b3'}
                    ),
                ], style={'width': '30%', 'display': 'inline-block'}),
                html.Div([
                    dcc.Dropdown(
                        id='position-dropdown',
                        style={'width': '100%', 'color': '#0056b3'}
                    ),
                ], style={'width': '50%', 'display': 'inline-block', 'margin-left': '10px'}),
            ], style={'width': '100%', 'margin': 'auto', 'text-align': 'left'}),
            html.Div(id='team-content')
        ]),
        dbc.Tab(label='Spain vs England Analysis', children=[
            dbc.Container([
                dbc.Row([
                    dbc.Col(dcc.Graph(figure=fig_age), md=6),
                    dbc.Col(dcc.Graph(figure=fig_market_caps), md=6)
                ]),
                dbc.Row([
                    dbc.Col(dcc.Graph(figure=fig_goals_position_spain), md=6),
                    dbc.Col(dcc.Graph(figure=fig_goals_position_england), md=6)
                ]),
                dbc.Row([
                    dbc.Col(dcc.Graph(figure=fig_goals_by_team), md=6),
                    dbc.Col(dcc.Graph(figure=fig_bubble), md=6)
                ]),
                dbc.Row([
                    dbc.Col(dcc.Graph(figure=fig_goals_conceded), md=6),
                    dbc.Col(dcc.Graph(figure=fig_passes), md=6)
                ]),
            ], fluid=True)  
        ])
    ])
], style={'backgroundColor': '#f0f8ff', 'color': '#0056b3', 'height': '150vh'})

@app.callback(
    Output('position-dropdown', 'options'),
    Input('team-dropdown', 'value')
)
def update_position_dropdown(selected_team):
    if selected_team == 'spain':
        positions = spain_df['position'].unique()
    elif selected_team == 'england':
        positions = england_df['position'].unique()
    return [{'label': pos, 'value': pos} for pos in positions]

@app.callback(
    Output('position-dropdown', 'value'),
    Input('position-dropdown', 'options')
)
def set_position_value(options):
    return options[0]['value'] if options else None

@app.callback(
    Output('team-content', 'children'),
    [Input('team-dropdown', 'value'),
     Input('position-dropdown', 'value')]
)
def update_content(selected_team, selected_position):
    if selected_team == 'spain':
        filtered_df = spain_df if selected_position is None else spain_df[spain_df['position'] == selected_position]
        fig = update_layout(px.scatter(filtered_df, x='age', y='marketvalue'), 'Market Value by Age', xaxis_title='Age', yaxis_title='Market Value', height=500, width=700)
        return html.Div([
            dash_table.DataTable(
                id='tablespain',
                columns=[
                    {'name': 'Name', 'id': 'name'},
                    {'name': 'Position', 'id': 'position'},
                    {'name': 'Age', 'id': 'age'},
                    {'name': 'Appearance', 'id': 'caps'},
                    {'name': 'Goals', 'id': 'goals'}
                ],
                data=filtered_df.to_dict('records'),
                style_table={'height': '300px', 'overflowY': 'auto'},
                style_header={'backgroundColor': 'white', 'color': 'black'},
                style_cell={'backgroundColor': 'white', 'color': 'black'}
            ),
            dcc.Graph(
                id='chart-spain',
                figure=fig
            )
        ])
    elif selected_team == 'england':
        filtered_df = england_df if selected_position is None else england_df[england_df['position'] == selected_position]
        fig = update_layout(px.scatter(filtered_df, x='age', y='marketvalue'), 'Market Value by Age', xaxis_title='Age', yaxis_title='Market Value', height=500, width=700)
        return html.Div([
            dash_table.DataTable(
                id='tableengland',
                columns=[
                    {'name': 'Name', 'id': 'name'},
                    {'name': 'Position', 'id': 'position'},
                    {'name': 'Age', 'id': 'age'},
                    {'name': 'Appearance', 'id': 'caps'},
                    {'name': 'Goals', 'id': 'goals'}
                ],
                data=filtered_df.to_dict('records'),
                style_table={'height': '300px', 'overflowY': 'auto'},
                style_header={'backgroundColor': 'white', 'color': 'black'},
                style_cell={'backgroundColor': 'white', 'color': 'black'}
            ),
            dcc.Graph(
                id='chart-england',
                figure=fig
            )
        ])

if __name__ == '__main__':
    app.run_server(debug=True, port=8070)
```
